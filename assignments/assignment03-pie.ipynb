```python
# import pandas as pd
import pandas as pd
```

```python
# pass the url directly into pandas read_csv function as stated in the documentation (https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html) 
df = pd.read_csv('https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download')
```

```python
# check out what the data looks like by showing the first 5 rows
df.head()
```

```python
# create a column called 'domain' that contains the domain part of the email address by splitting the 'Email' column at the '@' character and keeping the second part
# see: https://pandas.pydata.org/docs/reference/api/pandas.Series.str.split.html 
df['domain'] = df['Email'].str.split('@').str[1]

# check what that looks like
df.head()
```

```python
# basic code on making pie chart found: https://www.geeksforgeeks.org/python/how-to-create-pie-chart-from-pandas-dataframe/ 
# doucmentation on other parameters you can add: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html 
from matplotlib import pyplot as plt

# adjust the size of the pie chart
plt.figure(figsize=(7, 7))

# plot pie chart. group by domain and count the size of each group, format percentages, remove labels from slices, name the chart, explode the slices a bit, and select the colours.     
df.groupby('domain').size().plot(kind='pie', autopct='%1.0f%%', labels=None, title ='Email Domain Distribution', explode=[0.05,0.05,0.05], colors=['lightblue', 'lightgreen', 'lightcoral'])

# plot a legend (https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html) 
plt.legend(
    labels=df['domain'].unique(),    # unique domain names
    title='Domain',
    loc='best')
plt.show()
```
